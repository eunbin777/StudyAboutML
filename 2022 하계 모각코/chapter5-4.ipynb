```python
## 특정 단어를 포함한 게시글 찾기

# 데이터 준비하기
import pandas as pd
raw_total = pd.read_excel('../2022 하계 모각코 자료/5_Jeju_Hotplace/files/1_crawling_raw.xlsx')
raw_total.head()
```

```python
# 단어 선택하기
select_word = '오름'

check_list = []
for content in raw_total['content']:
    if select_word in str(content):
        check_list.append(True)
    else:
        check_list.append(False)

select_df = raw_total[check_list]
select_df.head()
```

```python
for i in select_df.index:
    print(select_df.loc[i, 'content'])
    print('-'*10)
select_df.head()
```

```python
fpath = f'../2022 하계 모각코 자료/5_Jeju_Hotplace/files/4_select_data_{select_word}.xlsx'
select_df.to_excel(fpath)
```

```python
select_word_list = ['오름', '박물관', '힐링', '섭지코지', '쇠소깍']

for select_word in select_word_list:
    check_list = [ ]
    for content in raw_total['content']:
        if select_word in str(content):
            check_list.append(True)
        else:
            check_list.append(False)
    
    select_df = raw_total[check_list]
    fpath = f'../2022 하계 모각코 자료/5_Jeju_Hotplace/files/4_select_data_{select_word}.xlsx'
    select_df.to_excel(fpath)
```
