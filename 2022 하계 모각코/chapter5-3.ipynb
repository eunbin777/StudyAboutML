```python
# 카카오 로컬 API를 활용한 장소 검색 해보기
from wsgiref import headers
import requests

searching = '합정 스타벅스'
url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(searching)

headers = {
    "Authorization" : "KakaoAK <카카오API 키>"
}

places = requests.get(url, headers = headers).json()['documents']
places
```

```python
def find_places(searching):
    # 접속 URL 만들기
    url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(searching)

    # headers 입력
    headers = {
        "Authorization": "KakaoAK <카카오API 키>"
    }

    # API 요청 및 정보 받기
    places = requests.get(url, headers = headers).json()['documents']

    # 필요한 정보 선택
    place = places[0]
    name = place['place_name']
    x = place['x']
    y = place['y']
    data = [name, x, y, searching]

    return data
```

```python
data = find_places('제주공항')
data
```

```python
# 반복작업 진행시 진행바 표시하기위한 라이브러리 tqdm 활용하기
from tqdm.notebook import tqdm
```

```python
import time

locations_inform = [ ]

for location in tqdm(locations):
    try:
        data = find_places(location)
        locations_inform.append(data)
        time.sleep(0.5)
    except:
        pass

locations_inform
```

```python
locations_inform_df = pd.DataFrame(locations_inform)
locations_inform_df.columns = ['카카오위치명', '경도', '위도', '인스타위치명']
locations_inform_df.to_excel('../2022 하계 모각코 자료/5_Jeju_Hotplace/files/3_locations.xlsx', index = False)
```

```python
location_counts_df = pd.read_excel('../2022 하계 모각코 자료/5_Jeju_Hotplace/files/3_location_counts.xlsx', index_col = 0)
locations_inform_df = pd.read_excel('../2022 하계 모각코 자료/5_Jeju_Hotplace/files/3_locations.xlsx')
```

```python
location_data = pd.merge(locations_inform_df, location_counts_df,
                        # inner로 양쪽 모두 있는 데이터만 포함
                        # 왼쪽에 위치한 데이터에서 해당 칼럼을 기준으로 병합
                        # 오른쪽에 위치한 데이터에서 인덱스를 기준으로 병합
                        how = 'inner', left_on = '카카오위치명', right_index = True)
location_data.head()
```

```python
location_data['카카오위치명'].value_counts()
```

```python
location_data = location_data.pivot_table(index = ['카카오위치명', '경도', '위도'], values = 'place', aggfunc = 'sum')
location_data.head()
```

```python
# 위치 정보 데이터를 지도에 표시할 수 있게 해주는 라이브러리
import folium

Mt_Hanla = [33.362500, 126.533694]
# 지도를 생성하는 명령어 folium.Map
# location 옵션으로 지도의 중심을 지정
map_jeju = folium.Map(location = Mt_Hanla, zoom_start = 11)

for i in range(len(location_data)):
    name = location_data['카카오위치명'][i]
    count = location_data['place'][i]
    size = int(count) * 5
    long = float(location_data['위도'][i])
    lat = float(location_data['경도'][i])
    folium.CircleMarker((long, lat), radius = size, color = 'red', popup = name).add_to(map_jeju)

map_jeju
```

```python
map_jeju.save('../2022 하계 모각코 자료/5_Jeju_Hotplace/files/3_jeju.html')
```

```python
from folium.plugins import MarkerCluster

locations = [ ]
names = [ ]

for i in range(len(location_data)):
    data = location_data.iloc[i]
    locations.append((float(data['위도']), float(data['경도'])))
    names.append(data['카카오위치명'])

Mt_Hanla = [33.362500, 126.533694]
map_jeju2 = folium.Map(location = Mt_Hanla, zoom_start= 11)

marker_cluster = MarkerCluster(
    locations = locations, popups = names,
    name = 'Jeju',
    overlay = True,
    control = True,
)

marker_cluster.add_to(map_jeju2)
folium.LayerControl().add_to(map_jeju2)

map_jeju2
```
