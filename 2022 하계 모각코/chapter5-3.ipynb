```python
def find_places(searching):
    # ì ‘ì† URL ë§Œë“¤ê¸°
    url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(searching)

    # headers ì…ë ¥
    headers = {
        "Authorization": "KakaoAK <ì¹´ì¹´ì˜¤API í‚¤>"
    }

    # API ìš”ì²­ ë° ì •ë³´ ë°›ê¸°
    places = requests.get(url, headers = headers).json()['documents']

    # í•„ìš”í•œ ì •ë³´ ì„ íƒ
    place = places[0]
    name = place['place_name']
    x = place['x']
    y = place['y']
    data = [name, x, y, searching]

    return data
```

```python
data = find_places('ì œì£¼ê³µí•­')
data
```

```python
# ë°˜ë³µì‘ì—… ì§„í–‰ì‹œ ì§„í–‰ë°” í‘œì‹œí•˜ê¸°ìœ„í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ tqdm í™œìš©í•˜ê¸°
from tqdm.notebook import tqdm
```

```python
import time

locations_inform = [ ]

for location in tqdm(locations):
    try:
        data = find_places(location)
        locations_inform.append(data)
        time.sleep(0.5)
    except:
        pass

locations_inform
```

```python
locations_inform_df = pd.DataFrame(locations_inform)
locations_inform_df.columns = ['ì¹´ì¹´ì˜¤ìœ„ì¹˜ëª…', 'ê²½ë„', 'ìœ„ë„', 'ì¸ìŠ¤íƒ€ìœ„ì¹˜ëª…']
locations_inform_df.to_excel('../2022 í•˜ê³„ ëª¨ê°ì½” ìë£Œ/5_Jeju_Hotplace/files/3_locations.xlsx', index = False)
```

```python
location_counts_df = pd.read_excel('../2022 í•˜ê³„ ëª¨ê°ì½” ìë£Œ/5_Jeju_Hotplace/files/3_location_counts.xlsx', index_col = 0)
locations_inform_df = pd.read_excel('../2022 í•˜ê³„ ëª¨ê°ì½” ìë£Œ/5_Jeju_Hotplace/files/3_locations.xlsx')
```

```python
location_data = pd.merge(locations_inform_df, location_counts_df,
                        # innerë¡œ ì–‘ìª½ ëª¨ë‘ ìˆëŠ” ë°ì´í„°ë§Œ í¬í•¨
                        # ì™¼ìª½ì— ìœ„ì¹˜í•œ ë°ì´í„°ì—ì„œ í•´ë‹¹ ì¹¼ëŸ¼ì„ ê¸°ì¤€ìœ¼ë¡œ ë³‘í•©
                        # ì˜¤ë¥¸ìª½ì— ìœ„ì¹˜í•œ ë°ì´í„°ì—ì„œ ì¸ë±ìŠ¤ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë³‘í•©
                        how = 'inner', left_on = 'ì¹´ì¹´ì˜¤ìœ„ì¹˜ëª…', right_index = True)
location_data.head()
```

```python
location_data['ì¹´ì¹´ì˜¤ìœ„ì¹˜ëª…'].value_counts()
```

```python
location_data = location_data.pivot_table(index = ['ì¹´ì¹´ì˜¤ìœ„ì¹˜ëª…', 'ê²½ë„', 'ìœ„ë„'], values = 'place', aggfunc = 'sum')
location_data.head()
```

```python
# ìœ„ì¹˜ ì •ë³´ ë°ì´í„°ë¥¼ ì§€ë„ì— í‘œì‹œí•  ìˆ˜ ìˆê²Œ í•´ì£¼ëŠ” ë¼ì´ë¸ŒëŸ¬ë¦¬
import folium

Mt_Hanla = [33.362500, 126.533694]
# ì§€ë„ë¥¼ ìƒì„±í•˜ëŠ” ëª…ë ¹ì–´ folium.Map
# location ì˜µì…˜ìœ¼ë¡œ ì§€ë„ì˜ ì¤‘ì‹¬ì„ ì§€ì •
map_jeju = folium.Map(location = Mt_Hanla, zoom_start = 11)

for i in range(len(location_data)):
    name = location_data['ì¹´ì¹´ì˜¤ìœ„ì¹˜ëª…'][i]
    count = location_data['place'][i]
    size = int(count) * 5
    long = float(location_data['ìœ„ë„'][i])
    lat = float(location_data['ê²½ë„'][i])
    folium.CircleMarker((long, lat), radius = size, color = 'red', popup = name).add_to(map_jeju)

map_jeju
```

```python
map_jeju.save('../2022 í•˜ê³„ ëª¨ê°ì½” ìë£Œ/5_Jeju_Hotplace/files/3_jeju.html')
```

```python
from folium.plugins import MarkerCluster

locations = [ ]
names = [ ]

for i in range(len(location_data)):
    data = location_data.iloc[i]
    locations.append((float(data['ìœ„ë„']), float(data['ê²½ë„'])))
    names.append(data['ì¹´ì¹´ì˜¤ìœ„ì¹˜ëª…'])

Mt_Hanla = [33.362500, 126.533694]
map_jeju2 = folium.Map(location = Mt_Hanla, zoom_start= 11)

marker_cluster = MarkerCluster(
    locations = locations, popups = names,
    name = 'Jeju',
    overlay = True,
    control = True,
)

marker_cluster.add_to(map_jeju2)
folium.LayerControl().add_to(map_jeju2)

map_jeju2
```
