```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

# 주피터노트북에서 결과를 출력하도록 설정
%matplotlib inline

# 글꼴 설정
plt.rc('font', family='NanumBarunGothic')

# 마이너스 기호 출력
plt.rc('axes', unicode_minus=False)

# 분석에 문제가 없는 경고 메세지는 숨김
import warnings
warnings.filterwarnings('ignore')
```

```python
train_data = train.copy()
```

```python
sns.displot(train_data['quality'], kde=False, bins=10)
plt.axis([0, 10, 0, 2500]) # [x축 최솟값, x축 최댓값, y축 최솟값, y축 최댓값]
plt.title("와인 품질")
plt.show()
```
