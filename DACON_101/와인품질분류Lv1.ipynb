```python
# 이상치 탐지 방법
# 대표적인 방법은 IQR(Inter Qunatile Range)로 사분위 값의 편차를 이용하여 이를 boxplot 그래프로 볼 수 있음
sns.boxplot(data=train['fixed acidity'])
```

```python
import numpy as np

# 'fixed acidity'가 25%인 값
quantile_25 = np.quantile(train['fixed acidity'], 0.25)
```

```python
# 'fixed acidity'가 75%인 값
quantile_75 = np.quantile(train['fixed acidity'], 0.75)
```

```python
# 둘의 차잇값
IQR = quantile_75 - quantile_25
```

```python
minimum = quantile_25 - 1.5 * IQR
maximum = quantile_75 + 1.5 * IQR

# 'fixed acidity'가 minimum보다 크고 maximum보다 작은 값들만 저장
train2 = train[(minimum <= train['fixed acidity']) & (train['fixed acidity'] <= maximum)]
train2.shape
```

```python
X = train.drop(columns = ['index', 'quality'])
Y = train['quality']
```

```python
kf = KFold(n_splits = 5, shuffle = True, random_state = 0)
```

```python
model = RandomForestClassifier(random_state = 0)
valid_scores = []
test_predictions = []
```

```python
from sklearn.metrics import accuracy_score
```

```python
for train_idx, valid_idx in kf.split(X,Y) : 
  X_tr = X.iloc[train_idx]
  Y_tr = Y.iloc[train_idx]

  X_val = X.iloc[valid_idx]
  Y_val = Y.iloc[valid_idx]

  model.fit(X_tr, Y_tr)
  
  valid_prediction = model.predict(X_val)
  score = accuracy_score(Y_val, valid_prediction)
  valid_scores.append(score)
  print(score)
print('평균 점수 : ', np.mean(valid_scores))
```

```python
for train_idx, valid_idx in kf.split(X, Y):
    X_tr = X.iloc[train_idx]
    Y_tr = Y.iloc[train_idx]

    X_val = X.iloc[valid_idx]
    Y_val = Y.iloc[valid_idx]

    model.fit(X_tr, Y_tr)

    test_prediction = model.predict(X_val)
    test_predictions.append(test_prediction)
```

```python
test_predictions = pd.DataFrame(test_predictions)
test_predictions
```

```python
test_prediction = test_predictions.mode()
test_prediction = test_predictions.values[0]
test_prediction
```

```python
sample_submission = pd.read_csv('./자료/와인/sample_submission.csv')
sample_submission['quality'] = test_prediction
sample_submission.to_csv('./자료/와인/submission_KFOLD.csv', index = False)
```
